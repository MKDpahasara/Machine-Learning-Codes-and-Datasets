# Decision Tree Regression and Random Forest Regression

Decision Tree Regression and Random Forest Regression are both machine learning algorithms employed for regression tasks, focusing on predicting continuous target variables. Decision trees construct a tree-like structure of decisions by recursively splitting the dataset based on features, with each leaf node representing a predicted outcome. However, decision trees can be prone to overfitting. On the other hand, Random Forest Regression addresses this issue by creating an ensemble of decision trees. Each tree is trained on a random subset of data and features, and their predictions are combined to improve generalization and reduce overfitting. Random Forests are often preferred in practice due to their robust performance across various datasets and their ability to handle more complex regression problems effectively.!



# Decision Tree Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset.head()
```

```python
X = dataset["Level"].values
y = dataset["Salary"].values
```

```python
X = X.reshape(len(X),1)
```

```python
y = y.reshape(len(y),1)
```

## Training the Decision Tree Regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

## Predicting a new result

```python
regressor.predict([[6.5]])
```

## Visualising the Decision Tree Regression results (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

## Predicting a new result

```python
regressor.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)


```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
