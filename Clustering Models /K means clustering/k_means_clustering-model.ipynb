# K-Means Clustering

K-means clustering is a fundamental unsupervised machine learning algorithm designed to partition a dataset into K distinct clusters based on the similarity of data points. Its iterative approach begins with the random initialization of K cluster centroids. In the assignment step, each data point is assigned to the nearest centroid, forming initial clusters. Subsequently, in the update step, centroids are recalculated by computing the mean of all data points assigned to each cluster. This process iterates until convergence, where centroids stabilize and no significant changes occur or a predefined number of iterations is reached. K-means aims to minimize the within-cluster sum of squares, essentially reducing the variance within clusters. Despite its simplicity, K-means finds applications in various domains including customer segmentation, image compression, and anomaly detection. However, it assumes clusters to be spherical and of similar sizes, and its performance can be affected by initialization and the number of clusters selected. Nonetheless, extensions and variations of K-means exist to address its limitations, making it a versatile and widely used clustering algorithm in the field of machine learning.



## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
